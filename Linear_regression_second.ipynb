```python
import numpy as np
np.random.seed(11)
```

```python
#prepare toy dataset

#use equation y= 2x1+5
x1 = np.random.randn(100)
y = 2*x1+5+np.random.randn(100)*0.9

#plot the graph
import matplotlib.pyplot as plt
plt.scatter(x1,y)
```

```python
x1 = (x1 - np.mean(x1))/np.std(x1)
plt.scatter(x1, y)
```

```python
#intitialise W and b
w = 1
b = 0

epochs = 100
alpha = 0.1
```

```python
def simple_linear_regression(w, b , x1):
    y_hat = w*x1+b
    return y_hat

#Compute RMSE
def RMSE(y_hat, y):
    return np.sqrt(np.sum(np.square(y_hat - y)))
```

```python
#Next tune the w and b based on loss
y_hat = simple_linear_regression(w, b, x1)
for epoch in range(epochs):
    loss = RMSE(y_hat, y)
    dw = np.sum(np.dot((y_hat -y), x1))/2
    db = np.sum(y_hat -y)/2

    #update w and b
    w -= alpha*dw/len(x1)
    b -= alpha*db/len(x1)

    y_hat = simple_linear_regression(w, b, x1)
    print("Loss in epoch {}: {}".format(epoch, RMSE(y_hat, y))) 
```
